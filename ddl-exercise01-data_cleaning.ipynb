```python
import pandas as pd
```

```python
#delimiter = seperate on
df1 = pd.read_csv('ddl-exercise01-steps_data.csv', delimiter=';')
df1.head()
```

```python
df2 = pd.read_csv('ddl-exercise01-survey_data.csv')
df2.head()
```

```python
df = pd.merge(df1, df2, on = 'id') #id is the common identifier
df.head()
```

```python
df['waist'].plot(kind = 'hist')
```

```python
df['waist'].value_counts().tail()
```

```python
#exclude extreme values
def high_low_waist(x):
    if(x > 130.0):
        return float('NaN')
    if(x < 40.0):
        return float ('NaN')
    else: 
        return x
df['waist'] = df['waist'].apply(high_low_waist)
```

```python
df['waist'].value_counts().tail()
```

```python
df['waist'].plot(kind = 'hist')
```
