```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df1 = pd.read_csv('movies.csv')
df1.head()
```

```python
df2 = pd.read_csv('ratings.csv')
df2.head()
```

```python
df = pd.merge(df1, df2, on = 'movieId')
df.head()
```

```python
pivot = pd.pivot_table(data=df, index='userId', columns='title', values='rating')
pivot.head()
```

```python
sorted_pivot = pivot.reindex(pivot.count().sort_values(ascending=False).index, axis=1)
sorted_pivot.head()
```

```python
pivot_full = sorted_pivot.fillna(0)
pivot_full.head()
```

```python
km = KMeans(n_clusters=5)
X = pivot_full.loc[:, 'Forrest Gump (1994)':"'71 (2014)"] #get the X variables from the dataframe
km = km.fit(X) #calculate the cluster centers
pivot_full['cluster'] = km.predict(X) #predict the clusters of each observation and store in the dataframe
pivot_full.head()
```

```python
pivot_full['cluster'].value_counts()
```

```python
pivot_full = pd.pivot_table(data = pivot_full, index = 'cluster')
```

```python
pivot_full = pivot_full.reindex(pivot.count().sort_values(ascending=False).index, axis=1)
pivot_full.head()
```

cluster 1: Cluster one doesn't seem to enjoy the action, war film, horror genre to much. They prefer the classics such as Forrest Gump. <br>
cluster 2: It looks likes the viewers of cluster two will most likely rate a movie they enjoyed. <br>
cluster 3: Viewers in cluster three don't enjoy sci-fi movies such as The Matrix and Star Wars, but they do enjoy classics such as The Shawshank Redemption. <br>
cluster 4: It looks like cluster four contains viewers who rate a movie when they don't particularly like it.
