```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind

df = pd.read_csv('ddl-exercise04-class_data.csv', delimiter=';')
df
```

```python
#independent variable = gender
#dependent variable = phones_owned
```

```python
#Make a plot of the relation showing the distribution for the two levels (e.g., histogram, violin plot, etc.)
sns.catplot(x='gender', y='phones_owned', kind='violin', data=df)
plt.show()
```

```python
female = df[df['gender']=='female']
male = df[df['gender']=='male'] 
```

```python
#The t-statistic and the p-value
ttest = ttest_ind(male['phones_owned'], female['phones_owned'], nan_policy='omit')

ttest1 = ttest[0]
print(f'The t-statistic is {ttest1}')

pvalue = ttest[1]
print(f'The p-value is {pvalue}. This p-value shows that there is no significance between the gender and the number of phones owned.')
```
