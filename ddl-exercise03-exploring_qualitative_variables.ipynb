```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('ddl-exercise03-class_data.csv', delimiter=';')
df
```

```python
#independent variable = gender
#dependent variable = like_tech
```

```python
#A frequency table with frequencies of your independent variable
df['gender'].value_counts()
```

```python
#A suitable graph of your independent variable
df2 = df['gender']
sns.countplot(df2)
```

```python
#A crosstable with column percentages
df3 = pd.crosstab(index = df['gender'], columns = df['like_tech'], normalize ='columns')
df3
```

```python
#A simple bar graph (not stacked) of the relation between the variables
sns.catplot(x='gender', hue='like_tech', kind='count', data=df)
```
