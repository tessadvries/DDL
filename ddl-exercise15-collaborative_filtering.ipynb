```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df1 = pd.read_csv('movies.csv')
```

```python
df2 = pd.read_csv('ratings.csv')
```

```python
df = pd.merge(df1, df2, on = 'movieId')
df.head()
```

```python
pivot = pd.pivot_table(data=df, index='userId', columns='title', values='rating')
```

```python
sorted_pivot = pivot.reindex(pivot.count().sort_values(ascending=False).index, axis=1)
sorted_pivot.head()
```

```python
#sorted_pivot.dropna(subset = ['Forrest Gump (1994)']).head()
```

```python
corr = sorted_pivot.corrwith(sorted_pivot['Forrest Gump (1994)'], axis=0, drop=False)
corr.sort_values(ascending=False)[0:10]
```

```python
reduced = sorted_pivot.dropna(axis=1, thresh=100)
corr = reduced.corrwith(reduced['Forrest Gump (1994)'])
corr.sort_values(ascending=False)[0:10]
```
