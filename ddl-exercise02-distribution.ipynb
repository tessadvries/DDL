```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('ddl-exercise02-class_data.csv')
df
```

```python
#measures of variation and central tendency
mean_size = df['texts_yesterday'].mean()
median_size = df['texts_yesterday'].median()

print(f'The median filesize is: {median_size} texts')
print(f'The mean filesize is: {mean_size} texts')

df['texts_yesterday'].name = 'Texts Yesterday' #Labeling the variable for plotting
textsyesterday = df['texts_yesterday'].dropna() 

sns.distplot(textsyesterday, kde=False)
plt.title('Number of texts sent yesterday')
plt.axvline(median_size, 0, 100, color='green', label='median') #This adds a vertical line at x-position median_size, from y = 0 to y = 100 
plt.axvline(mean_size, 0, 100, color='blue', label='mean') 
plt.legend() #This adds a legend. It works automatically because I've set the labels in the previous lines
plt.show()
```

```python
#boxplot of instagram followers
df2 = df[df['instagram_followers']<500]

sns.boxplot(df2['instagram_followers']) 
plt.title('Instagram Followers')
plt.show()
```

```python
#histogram of linkedin connections
sns.distplot(df['linkedin_con'].dropna(), kde=False)
plt.title('LinkedIn Connections')
plt.xlabel('LinkedIn Connections')
plt.show()
```

```python
#density plot of phones owned
df['phones_owned'].name = 'Phones Owned' #Labeling the variable for plotting
phones = df['phones_owned'].dropna() #Selecting the phones owned column and dropping NA.
sns.distplot(phones) 
plt.title('Phones Owned')
plt.show()

#skew is on the left, this shows that it is more common to have owned a lower amount of phones. in the plot it can be seen that there is
#an outlier near the 25 on the x-as.
```

```python
#violin plot of amount of apps on phone
df3 = df[df['apps_phone']<150]
sns.violinplot(df3['apps_phone'])
sns.swarmplot(df3['apps_phone'], color='purple')
plt.title('Apps on Phone')
plt.show()
```
